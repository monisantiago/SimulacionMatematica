## Archivo datosCT.xlsx

```python
data = pd.read_excel("datosCT.xlsx")
```

```python
data
```

### Utilidad

```python
data["Utilidad"] = data["Ingreso"] - data["Gasto"]
```

```python
data
```

### Promedio de Gastos en los Meses que la Utilidad fue Negativa

```python
data["Meses en Perdida"] = data["Gasto"].where(data["Utilidad"]<0,0)
```

```python
data
```

```python
data["Meses en Perdida"].mean()
```
